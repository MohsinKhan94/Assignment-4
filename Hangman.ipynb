```python
import random

# List of words to choose from
words = ['python', 'java', 'javascript', 'ruby', 'cplusplus', 'html', 'css', 'react', 'nodejs']
word_to_guess = random.choice(words)
guessed_letters = []
attempts = 6  # Number of attempts before losing

# Hangman structure
hangman_pics = [
    '''
     ------
     |    |
          |
          |
          |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
          |
          |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
     |    |
          |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
    /|    |
          |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
    /|\\   |
          |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
    /|\\   |
    /     |
          |
          |
    =========
    ''',
    '''
     ------
     |    |
     O    |
    /|\\   |
    / \\   |
          |
          |
    =========
    '''
]

# Function to display the current state of the word
def display_word():
    return ''.join([letter if letter in guessed_letters else '_' for letter in word_to_guess])

print("Welcome to Hangman!")

# Game loop
while attempts > 0:
    print("\nWord to guess: ", display_word())
    print(f"Guessed letters: {', '.join(guessed_letters)}")
    print(f"You have {attempts} attempts left.")
    print(hangman_pics[6 - attempts])  # Display hangman based on the number of attempts

    # Get the user's guess
    guess = input("Guess a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a valid letter.")
        continue

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    if guess in word_to_guess:
        print(f"Good guess! '{guess}' is in the word.")
    else:
        attempts -= 1
        print(f"Oops! '{guess}' is not in the word.")

    # Check if the player has guessed all letters
    if all(letter in guessed_letters for letter in word_to_guess):
        print(f"\nCongratulations! You guessed the word: {word_to_guess}")
        break
else:
    print(f"\nYou lost! The word was: {word_to_guess}")
    print(hangman_pics[6])  # Display the final hangman image
```
